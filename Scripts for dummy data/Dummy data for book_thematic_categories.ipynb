```python
import random

book_genres = [
    "Mystery",
    "Romance",
    "Science Fiction",
    "Fantasy",
    "Thriller",
    "Horror",
    "Biography",
    "Self-Help",
    "Crime",
    "Adventure",
    "Comedy",
    "Drama",
    "Poetry",
    "Action",
    "Graphic Novel"
]

for book_id in range(1, 101):
    num_genres = random.randint(1, min(3, len(book_genres)))
    random_genres = random.sample(book_genres, num_genres)
    for genre in random_genres:
        print(f"('{genre}', {book_id}),")
```
